```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime, timedelta

# Define the date range
now = datetime.now()
start_date = now - timedelta(days=365 * 2)
end_date = now

# Download the data
nvda_data = yf.download('NVDA', start=start_date, end=end_date, interval="1d")
qqq_data = yf.download('QQQ', start=start_date, end=end_date, interval="1d")

# Earnings call dates for the past two years
earnings_dates = pd.to_datetime([
    "2021-02-24", "2021-05-26", "2021-08-18", "2021-11-17",
    "2022-02-16", "2022-05-25", "2022-08-24", "2022-11-16",
    "2023-02-22", "2023-05-24", "2023-08-23", "2023-11-21"
])

# Create a DataFrame with both closing prices
combined_data = pd.DataFrame({
    'NVDA': nvda_data['Close'],
    'QQQ': qqq_data['Close']
})

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(combined_data.index, combined_data['NVDA'], label='NVDA')
plt.plot(combined_data.index, combined_data['QQQ'], label='QQQ')

# Marking earnings call dates
for date in earnings_dates:
    if date in combined_data.index:
        plt.axvline(x=date, color='k', linestyle='--', alpha=0.7)

plt.title('NVIDIA vs QQQ Closing Prices with Earnings Call Dates')
plt.ylabel('Price (USD)')
plt.xlabel('Date')
plt.legend()
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime, timedelta

# Define the date range
now = datetime.now()
start_date = now - timedelta(days=365 * 2)
end_date = now

# Download the data for NVDA and SPY
nvda_data = yf.download('NVDA', start=start_date, end=end_date, interval="1d")
spy_data = yf.download('SPY', start=start_date, end=end_date, interval="1d")

# Earnings call dates for the past two years
earnings_dates = pd.to_datetime([
    "2021-02-24", "2021-05-26", "2021-08-18", "2021-11-17",
    "2022-02-16", "2022-05-25", "2022-08-24", "2022-11-16",
    "2023-02-22", "2023-05-24", "2023-08-23", "2023-11-21"
])

# Create a DataFrame with NVDA and SPY closing prices
combined_data = pd.DataFrame({
    'NVDA': nvda_data['Close'],
    'SPY': spy_data['Close']
})

# Calculate the ratio of NVDA price over SPY price
combined_data['NVDA/SPY'] = combined_data['NVDA'] / combined_data['SPY']

# Plotting NVDA, SPY, and the ratio
plt.figure(figsize=(12, 8))

# Subplot 1: NVDA and SPY prices
plt.subplot(2, 1, 1)
plt.plot(combined_data.index, combined_data['NVDA'], label='NVDA')
plt.plot(combined_data.index, combined_data['SPY'], label='SPY')
plt.title('NVIDIA vs SPY Closing Prices with Earnings Call Dates')
plt.ylabel('Price (USD)')
plt.legend()

# Marking earnings call dates on the first plot
for date in earnings_dates:
    if date in combined_data.index:
        plt.axvline(x=date, color='k', linestyle='--', alpha=0.7)

# Subplot 2: NVDA/SPY Ratio
plt.subplot(2, 1, 2)
plt.plot(combined_data.index, combined_data['NVDA/SPY'], label='NVDA/SPY Ratio', color='purple')
plt.title('NVDA/SPY Price Ratio')
plt.ylabel('Ratio')
plt.xlabel('Date')

# Marking earnings call dates on the ratio plot
for date in earnings_dates:
    if date in combined_data.index:
        plt.axvline(x=date, color='k', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
